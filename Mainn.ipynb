```python
# %%
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
def get_Price(newsoup):
    try:
        Price=newsoup.find('span', attrs={'class':'a-price-whole'}).text
        # print(Price)
    except AttributeError:
        Price="Error"
    return Price
```

```python
def get_Rating(newsoup):
    try:
        Rating=newsoup.find('span', attrs={'class':'a-icon-alt'}).text
    except AttributeError:
        Rating="Error"
    return Rating
```

```python
def get_Total_Reviews(newsoup):
    try:
        Total_Review=newsoup.find('span', attrs={'id':'acrCustomerReviewText'}).text
    except AttributeError:
        Total_Review="Error"
    return Total_Review
```

```python
def get_About(newsoup):
    try:
        feature_bullets= newsoup.find('div', {'id': 'feature-bullets'}).text
        About=feature_bullets.replace('\n','')
        if len(About)>30:
            About="EMPTY"
    except AttributeError:
        About="Error"
    return About
```

```python

# %%
URL = 'https://www.amazon.in/s?k=mobiles&i=electronics'

HEADERS = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36'})  
    
webpage=requests.get(URL,headers=HEADERS)
webpage

type(webpage.content)


soup = BeautifulSoup(webpage.content, "html.parser")

links = soup.find_all('a', class_='a-link-normal s-line-clamp-2 s-link-style a-text-normal')
print(len(links)) 
```

```python


links_list=[]
for link in links:
    links_list.append(link.get('href'))

d = {'Name':[],'Price':[],'Rating':[],'Total_Reviews':[],'About':[],'Link':[]}
#,'Link':[]
```

```python

#
Loop=0

for link in links_list:
#while Loop<8:
    new_webpage=requests.get("https://www.amazon.in" + link, headers=HEADERS, timeout=14)
    print("Loop=",Loop)
    Loop+=1
    Name=""
    newsoup = BeautifulSoup(new_webpage.content, "html.parser")
    print("https://www.amazon.in" + link)

    Name=get_Name(newsoup)
    Price=get_Price(newsoup)
    Rating=get_Rating(newsoup)
    Total_Reviews=get_Total_Reviews(newsoup)
    About=get_About(newsoup)

    print(Name)
    print(Price)
    print(Rating)
    print(Total_Reviews)
    print(About)
    
    d['Name'].append(Name)
    d['Price'].append(Price)
    d['Rating'].append(Rating)
    d['Total_Reviews'].append(Total_Reviews)
    d['About'].append(About)
    d['Link'].append("https://www.amazon.in"+link)
```

```python
amazon_df=pd.DataFrame.from_dict(d)
amazon_df['Name'].replace('','NaN', inplace=True)
amazon_df=amazon_df.dropna(subset=['Name'])
amazon_df.to_csv('amazon.csv',header=True,index=False)
```
